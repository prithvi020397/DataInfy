```python
rate = pd.DataFrame(one.groupby('title')['rating'].mean())
rate.head()
```

```python
rate['num of ratings'] = pd.DataFrame(one.groupby('title')['rating'].count())
rate.head()
```

```python
plt.figure(figsize=(10,4))
rate['rating'].hist(bins=40)
```

```python
moviematrix = one.pivot_table(index='user_id',columns='title',values='rating')
moviematrix.head()
```

```python
rate.sort_values('num of ratings',ascending=False).head(10)
```

```python
rate.head()
```

```python
starwars_user_ratings = moviematrix['Star Wars (1977)']
dumb_user_ratings = moviematrix['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
dumb_user_ratings.head()
```

```python
similar_to_starwars = moviematrix.corrwith(starwars_user_ratings)
similar_to_liarliar = moviematrix.corrwith(dumb_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(8)
```

```python
corr_starwars = corr_starwars.join(rate['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(rate['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
